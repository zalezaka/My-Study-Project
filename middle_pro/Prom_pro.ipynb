# 2. Выяви самый сложный и самый простой экзамен: найди курсы и экзамены в рамках курса, которые обладают самой низкой и самой высокой завершаемостью. 
*завершаемость = кол-во успешных экзаменов / кол-во всех попыток сдать экзамен

```python
exams = df_st_full1.query("assessment_type == 'Exam'")\
                    .groupby(['code_module','id_assessment'], as_index = False)\
                    .agg({'id_student':'count'}).rename(columns = {'id_student':'counts'})

exams['suc'] = (df_st_full1.query("score >= 40 & assessment_type == 'Exam'")\
                           .groupby(['code_module','id_assessment'], as_index = False)\
                           .agg({'id_student':'count'})).id_student

exams['dif_ex'] = exams['suc'] / exams['counts'] 

exams.sort_values('dif_ex')

#ответ - 25361 самый легкий Экзамен и 25340 самый сложный экзамен
```

# 6. RFM-сегментация 

Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй адаптированные RFM-кластеры студентов, чтобы качественно оценить свою аудиторию. В адаптированной кластеризации можешь выбрать следующие метрики: R - среднее время сдачи одного экзамена, F - завершаемость курсов, M - среднее количество баллов, получаемое за экзамен. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Описание подхода можно найти тут. (35 баллов)

Для решения задачи проведи предварительное исследование данных и сформулируй, что должно считаться курсом. Обосновать свой выбор ты можешь с помощью фактов сдачи экзаменов, распределения студентов и уникальный идентефикатор курса.



```python
from datetime import timedelta
import matplotlib.pyplot as plt
```

Каждый студент характеризуется тремя параметрами: R как быстро он сдал экзамены, F сколько курсов из всех он хамершал (завершаемость), M  какая средняя оценка у него по всем предметам

R - среднее время сдачи одного экзамена
M - среднее количество баллов

```python
rfm = df_st_full2.groupby('id_student', as_index = False).agg({'date_submitted':'mean','score':'mean'})
```

F - завершаемость курсов

```python
a = df_st_full2.query("score>=40 & assessment_type == 'Exam'").groupby('id_student',as_index=True ).agg({'id_assessment':'nunique'})
b = df_st_full2.query("assessment_type == 'Exam'").groupby('id_student', as_index= True ).agg({'id_assessment':'count'})
df = pd.DataFrame(a['id_assessment']/ b['id_assessment'])
rfm = rfm.merge(df, on = 'id_student', how = 'inner').rename(columns = {'date_submitted':'recency',
                                                                 'score':'monetary',
                                                                 'id_assessment':'frequency',
    
})
```

```python
quintiles = rfm[['recency', 'monetary']].quantile([.2, .4, .6, .8]).to_dict()
```

```python
def r_score(x):
    if x <= quintiles['recency'][.2]:
        return 5 #быстро сдал экзамены
    elif x <= quintiles['recency'][.4]:
        return 4
    elif x <= quintiles['recency'][.6]:
        return 3
    elif x <= quintiles['recency'][.8]:
        return 2
    else:
        return 1 #долго сдавал экзамены

def m_score(x):
    if x <= quintiles['monetary'][.2]:
        return 1 
    elif x <= quintiles['monetary'][.4]:
        return 2
    elif x <= quintiles['monetary'][.6]:
        return 3
    elif x <= quintiles['monetary'][.8]:
        return 4
    else:
        return 5 #плохо сдавал экзамены
    
    
def fm_score(x):
    if x==0.5:
        return 0 #не сдана половина
    else: 
        return 1
 
        
 
```

```python
rfm['R'] = rfm['recency'].apply(lambda x: r_score(x))
rfm['M'] = rfm['monetary'].apply(lambda x: m_score(x))
rfm['F'] = rfm['frequency'].apply(lambda x: fm_score(x))
```

```python
rfm['RFM_Score'] = rfm['R'].map(str) + rfm['M'].map(str) + rfm['F'].map(str)
rfm.head(10)
```

```python
segt_map = {
    r'[1-5][1-5][0]': 'на вылет',
    r'[3-5][3-5][0]': 'оценки ок, но половина курсов не сдана',
    r'[1-2][1-2][1]': 'плохие оценки',
    r'[3-5][3-5][1]': 'отличники',
    r'[3-5][1-2][1]': 'плохие оценки, но куда-то торопится с экзаменами',
    r'[1-2][3-5][1]': 'хорош но долго сдает экзамены'  
}

rfm['Segment'] = rfm['RFM_Score'].replace(segt_map, regex=True)
rfm.head(10)
```

```python

segments_counts = rfm['Segment'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='red')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)


plt.show()
```
